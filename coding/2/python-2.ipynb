# Can you generate code for me to plot a gaussian distribution in jupyter notebooks?


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the distribution parameters
mu = 0
sigma = 1

# Create the data
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)

# Plot the distribution
plt.plot(x,y)
plt.title('Gaussian Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# Can you generate code to illustrate the concept of the law of large numbers in jupyter notebooks? I'm having some trouble contextualizing the idea

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the distribution parameters
mu = 0
sigma = 1

# Generate random samples from a normal distribution
n = 1000
samples = np.random.normal(mu, sigma, n)

# Calculate the cumulative mean of the samples
cumulative_mean = np.cumsum(samples) / np.arange(1, n+1)

# Plot the cumulative mean as a function of sample size
plt.plot(cumulative_mean)
plt.axhline(y=mu, color='r', linestyle='-')
plt.title('Law of Large Numbers')
plt.xlabel('Sample Size')
plt.ylabel('Cumulative Mean')
plt.show()
```

# I'm interested in generating an animated plot for how the law of large numbers works in python. Could you help me with this?


```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

def plot_wlln(trials):
    """Plot the weak law of large numbers for a single die."""
    # Generate random rolls of a die
    rolls = np.random.randint(1, 7, trials)
    # Compute the cumulative average of the rolls
    averages = np.cumsum(rolls) / np.arange(1, trials + 1)
    # Plot the cumulative averages
    fig, ax = plt.subplots()
    ax.plot(averages)
    ax.axhline(y=3.5, color='r', linestyle='-')
    ax.set_xlabel('Number of Rolls')
    ax.set_ylabel('Average')
    ax.set_title('Weak Law of Large Numbers')
    return fig

trials = st.slider("Number of dice rolls", 10, 25000)
st.pyplot(plot_wlln(trials))
```
